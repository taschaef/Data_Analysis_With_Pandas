Trends from Analysis
-------------------------------
1. Small and medium schools noticably outpreformed large schools in math and reading; on average small and medium schools had at least 10% more students passing math, reading or both. Small & medium schools scored around 93.5% for percentage of students passing math, while large schools were only at 69.9%. Similarly, small & medium schools had approximately 96.3% of their students passing reading, and large schools only had about 82% of their students passing reading. When we consider both percentages is where we can easily see the noticable difference; when percentage of students passing both math and reading are considered smaller and medium schools significantly outpreform larger schools. Small and medium schools clock in around 90% for students passing math and reading whereas larger schools only have about 58% of their students passing both. This suggests that smaller and medium schools may be able to work more closely with their students as the teacher:student ratio is lower, and this in turn may produce better scores. There is much else to be analyzed before this can be definitively declared. Smaller schools might potentially have more money to spend per student, so rather than a teacher:student ratio, it could in fact be due to a money:student ratio. 

2. The average math score and the average reading scores tend to decrease as more money is spent on students individually. When the least amount of money is spent per student, less than 585, the average math score across schools is at its highest, about 83.4. The average math score is at its lowest (69.9%) when the most amount of money (645-680) is spent per student. The same trend can be observed in the average reading score. When less than 585 is spent per student the average reading score is 83.9, but drops to 81.0 when spending is between 645-680. This suggests that there is more to student success (i.e. math and reading scores) than just the amount of money spent per student. More evidence will need to be examined before this is proved conclusively. For example, do schools with less money have to rely more on study groups than practice booklets? Could this play a significant role in test scores as the learning culture might be more interactive and more of a community event?  

3. By looking at the tables depicting the highest performing and lowest performing schools based on percentage of students passing both math and reading, it is obvious charter schools outpreform district schools. The top eight schools in the district are charter schools, with district schools only ranking on the highest performing school table at the eight spot. District schools make up the seven lowest performing schools. This could be due to several factors; charter schools often serve middle to upper-middle class neighborhoods, whereas district schools often do not have the same socioeconomic benefits. Charter schools, on average, offer better pay & benefits to their instructors whereas district instructors are often financed by taxes/federal government; so it is possible for charter schools to obtain higher-quality instructors than their district counterparts based on pay offered alone.
